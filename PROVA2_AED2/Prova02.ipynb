Dado o código a seguir, que contém uma implementação incompleta e com falhas do algoritmo de Selection Sort,

```python
# implementação selection sort c/ falhas ...

def selection_sort(arr):
    for i in range(len(arr)-1):
        min_idx = i + 1
        for j in range(i, len(arr)):
            if arr[j] < arr [i]:
                min_idx = j
            arr [min_idx] = arr[i]
            arr[i] = arr[min_idx]
        return arr
```

**a) Identifique e corrija os erros presentes no código.**

```python
# implementação selection sort corrigida

def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

**Original: min_idx = i + 1**  
**Correção: min_idx = i**  
O código começava a procurar o menor número a partir do segundo elemento, quando deveria começar do primeiro, o i atual.  

**Original: for j in range(i, len(arr))**   
**Correção: for j in range(i + 1, n)**  
O loop interno do código comparava elementos que não havia necessidade. O correto é começar a comparar a partir do próximo elemento.  

**Original: if arr[j] < arr [i]**  
**Correção: if arr[j] < arr[min_idx]**  
Essa linha de código comparava o número da posição i ao invés de comparar com o menor número encontrado até o momento.  

**Original: arr [min_idx] = arr[i] / arr[i] = arr[min_idx]**  
**Correção: arr[i], arr[min_idx] = arr[min_idx], arr[i]**  
A troca de valores ao ser realizada, perdia-se um valor, além de ocorrer dentro do loop interno.  
O correto é essa linha estar fora do loop interno, depois que o min_idx final foi encontrado.  

**Original: return arr (dentro do loop)**  
**Correção: return arr (fora do loop)**  
Aa função parava de ser executada antes de terminar de ordenar a lista. Se o return é colodado no final da função, depois de todos os loops externos terminara, o código será executado corretamente como deveria.

**d) Imprima também o maior elemento da lista ordenada SEM usar max() nem sorted().**

```python
lista = [53, 39, 81, 22, 11, 60, 93]

def find_max(lista_ordenada):
    max_element = lista_ordenada[0]
    for element in lista_ordenada:
        if element > max_element:
            max_element = element
    return max_element

print("Maior elemento:", find_max(lista))
```
